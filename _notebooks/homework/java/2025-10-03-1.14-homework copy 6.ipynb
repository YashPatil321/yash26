## Popcorn  2

```java
public class Counter {
    private int count;

    public void add(int x) {
        count += x;
    }

    public void subtract(int x) {
        count -= x;
    }

    public void multiply(int x) {
        count *= x;
    }

    public void divide(int x) {
        count /= x;
    }

    public int getCount() {
        return count;
    }

    public static void main(String[] args) {
        Counter c = new Counter();
        c.add(10);
        c.subtract(2);
        c.multiply(3);
        c.divide(4);
        System.out.println("Final count: " + c.getCount());
    }
}

Counter.main(null); // Run the counter simulation
```

## Homework  

```java
import java.util.ArrayList;

public class Student {
    // Instance Variables
    private String name;
    private int totalPoints;
    private int numAssignments;
    private ArrayList<Integer> grades; // stores all grades

    // Constructor
    public Student(String name) {
        this.name = name;
        this.totalPoints = 0;
        this.numAssignments = 0;
        this.grades = new ArrayList<>();
    }

    // Adds a grade and updates totals
    public void addGrade(int points) {
        if (points < 0 || points > 100) {
            System.out.println("Invalid grade. Must be between 0 and 100.");
            return;
        }
        grades.add(points);
        totalPoints += points;
        numAssignments++;
    }

    // Returns average
    public double getAverage() {
        if (numAssignments == 0) return 0.0;
        return (double) totalPoints / numAssignments;
    }

    // Returns letter grade
    public String getLetterGrade() {
        double avg = getAverage();
        if (avg >= 90) return "A";
        else if (avg >= 80) return "B";
        else if (avg >= 70) return "C";
        else if (avg >= 60) return "D";
        else return "F";
    }

    // Returns highest grade
    public int getHighestGrade() {
        if (grades.isEmpty()) return 0;
        int max = grades.get(0);
        for (int grade : grades) {
            if (grade > max) max = grade;
        }
        return max;
    }

    // Returns lowest grade
    public int getLowestGrade() {
        if (grades.isEmpty()) return 0;
        int min = grades.get(0);
        for (int grade : grades) {
            if (grade < min) min = grade;
        }
        return min;
    }

    // Drops lowest grade
    public void dropLowestGrade() {
        if (grades.isEmpty()) return;
        int lowest = getLowestGrade();
        grades.remove(Integer.valueOf(lowest));
        totalPoints -= lowest;
        numAssignments--;
    }

    // Compares averages between two students
    public String compareTo(Student other) {
        double myAvg = getAverage();
        double otherAvg = other.getAverage();
        if (myAvg > otherAvg)
            return name + " has a higher average than " + other.name + ".";
        else if (myAvg < otherAvg)
            return name + " has a lower average than " + other.name + ".";
        else
            return name + " and " + other.name + " have the same average.";
    }

    // Resets all data
    public void reset() {
        grades.clear();
        totalPoints = 0;
        numAssignments = 0;
    }

    // Prints report
    public void printReport() {
        System.out.println("Student: " + name);
        System.out.println("Total Points: " + totalPoints);
        System.out.println("Assignments: " + numAssignments);
        System.out.printf("Average: %.2f\n", getAverage());
        System.out.println("Letter Grade: " + getLetterGrade());
        System.out.println("Highest Grade: " + getHighestGrade());
        System.out.println("Lowest Grade: " + getLowestGrade());
        System.out.println("--------------------------");
    }

    // Main method (test runner)
    public static void main(String[] args) {
        Student s1 = new Student("Yash");
        Student s2 = new Student("Ava");

        s1.addGrade(95);
        s1.addGrade(87);
        s1.addGrade(92);

        s2.addGrade(78);
        s2.addGrade(84);
        s2.addGrade(69);

        s1.printReport();
        s2.printReport();

        s2.dropLowestGrade();
        System.out.println("After dropping lowest grade for Ava:");
        s2.printReport();

        System.out.println(s1.compareTo(s2));

        s1.reset();
        System.out.println("After resetting Yashâ€™s record:");
        s1.printReport();
    }
}

// run
Student.main(null); // Run the student grade management system
```
