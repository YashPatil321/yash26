## Hack 1

```java
class Bro {
    String name;
    static String catchphrase = "Sup bro?";

    public Bro(String name) {
        this.name = name;
    }

    // Instance method — needs an object
    public void sayHi() {
        System.out.println("Hey, I'm " + name);
    }

    // Static (class) method — belongs to the class
    public static void sayCatchphrase() {
        System.out.println(catchphrase);
    }

    public static void main(String[] args) {
        // Works: calling static method directly
        sayCatchphrase();

        // Error fixed: must use an object to call instance method
        Bro alex = new Bro("Alex");
        alex.sayHi();

        // Works both ways (object can access static methods too)
        alex.sayCatchphrase();
    }
}

Bro.sayCatchphrase(); // Works: calling static method directly
```

## HOMEWORK HACK

```java
class Robot {
    // Instance variables
    String name;
    int power;
    int health;
    String weapon; // your custom variable!

    // Static variable
    static double fightDuration = 0;

    // Constructor
    public Robot(String name, int power, int health, String weapon) {
        this.name = name;
        this.power = power;
        this.health = health;
        this.weapon = weapon;
    }

    // Instance method: attack
    public void attack(Robot opponent) {
        System.out.println(name + " attacks " + opponent.name + " with " + weapon + "!");
        opponent.health -= power;
        if (opponent.health < 0) opponent.health = 0;
    }

    // Instance method: print status
    public void printStatus() {
        System.out.println(name + " | Power: " + power + " | Health: " + health + " | Weapon: " + weapon);
    }

    // Class method: determine which fighter is stronger
    public static Robot strongerFighter(Robot r1, Robot r2) {
        if (r1.power > r2.power) return r1;
        else if (r2.power > r1.power) return r2;
        else return null; // equal strength
    }

    // Class method: start the battle
    public static void beginBattle(Robot r1, Robot r2) {
        System.out.println("\n⚔️  BATTLE STARTS BETWEEN " + r1.name + " AND " + r2.name + "! ⚔️");
        fightDuration = Math.random() * 10 + 5; // random duration for fun

        while (r1.health > 0 && r2.health > 0) {
            r1.attack(r2);
            r2.attack(r1);
            System.out.println();
            r1.printStatus();
            r2.printStatus();
            System.out.println("------");
        }

        System.out.println("\nBattle lasted " + String.format("%.2f", fightDuration) + " minutes.");
        if (r1.health > 0 && r2.health <= 0)
            System.out.println(r1.name + " wins the battle!");
        else if (r2.health > 0 && r1.health <= 0)
            System.out.println(r2.name + " wins the battle!");
        else
            System.out.println("It’s a draw!");
    }

    // Main method
    public static void main(String[] args) {
        Robot alpha = new Robot("Alpha", 30, 100, "Laser Cannon");
        Robot beta = new Robot("Beta", 25, 110, "Plasma Blade");

        System.out.println("=== ROBOT STATUS BEFORE BATTLE ===");
        alpha.printStatus();
        beta.printStatus();

        System.out.println("\nStronger fighter: " + (strongerFighter(alpha, beta) != null 
            ? strongerFighter(alpha, beta).name 
            : "Equal Strength"));

        beginBattle(alpha, beta);
    }
}

Robot.main(null); // Run the battle simulation
```
