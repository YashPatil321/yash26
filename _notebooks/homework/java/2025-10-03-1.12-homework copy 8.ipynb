## Hack 1

```java
class Book {
    String title;
    int pages;

    // Method to print book info
    public void printInfo() {
        System.out.println("Title: " + title + ", Pages: " + pages);
    }
}

class MainPopcorn {
    public static void main(String[] args) {
        Book myBook = new Book();
        myBook.title = "The Lightning Thief";
        myBook.pages = 377;

        myBook.printInfo();
    }
}

// Run the class
MainPopcorn.main(null);
```

Class vs. Object:
A class is a blueprint that defines how objects are made, while an object is an actual instance created from that blueprint.

Reference variable:
It stores the memory address (reference) of an object, not the object itself.

Method every object has:
toString() (also inherited from the Object class).

## HOMEWORK HACK 

```java
public class Student {
    // Instance variables
    String name;
    int grade;
    int pets;
    int siblings;
    String hobby; // custom variable

    // Constructor
    public Student(String name, int grade, int pets, int siblings, String hobby) {
        this.name = name;
        this.grade = grade;
        this.pets = pets;
        this.siblings = siblings;
        this.hobby = hobby;
    }

    // Method to print info
    public void printInfo() {
        System.out.println("Name: " + name);
        System.out.println("Grade: " + grade);
        System.out.println("Pets: " + pets);
        System.out.println("Siblings: " + siblings);
        System.out.println("Hobby: " + hobby);
        System.out.println("-------------------------");
    }

    public static void main(String[] args) {
        // Create 3 imaginary students
        Student s1 = new Student("Liam", 10, 2, 1, "Soccer");
        Student s2 = new Student("Ava", 11, 0, 3, "Drawing");
        Student s3 = new Student("Noah", 9, 1, 2, "Gaming");

        // Print info for all
        System.out.println("=== Student List ===");
        s1.printInfo();
        s2.printInfo();
        s3.printInfo();

        // Reference variable example
        System.out.println("=== Nickname Reference ===");
        Student nickname = s1; // nickname refers to same student as s1
        nickname.name = "Lee"; // change name using reference

        // Print through both references
        System.out.println("Original (s1) Info:");
        s1.printInfo();

        System.out.println("Nickname (nickname) Info:");
        nickname.printInfo();
    }
}

Student.main(null); // Run the student class
```
