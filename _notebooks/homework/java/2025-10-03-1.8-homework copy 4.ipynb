## Hack 1

```java
/**
 * Calculates the sum of all positive even integers in the given array.
 *
 * <p>Precondition: The input array {@code nums} is not null.</p>
 * <p>Postcondition: Returns the total sum of all elements in {@code nums}
 * that are both positive and even. Does not modify the input array.</p>
 *
 * @param nums an array of integers to analyze
 * @return the sum of all positive even numbers in {@code nums}
 */
public int doSomething(int[] nums) {
    int result = 0;
    for (int i = 0; i < nums.length; i++) {
        if (nums[i] > 0 && nums[i] % 2 == 0) {
            result += nums[i];
        }
    }
    return result;
}
```

## Reflection 

1. Why is documentation more important in team projects than solo projects?
Because team projects involve multiple developers, documentation helps everyone understand the code’s purpose, logic, and usage without guessing. It saves time, prevents mistakes, and keeps the project consistent even when members change or new ones join.

2. Give an example of when a method SHOULD be documented and when it SHOULD NOT.
A method should be documented if it performs a key function, has complex logic, or is used by others (like an API method). It should not be documented if it’s private, simple, or self-explanatory (like a one-line getter or setter).
